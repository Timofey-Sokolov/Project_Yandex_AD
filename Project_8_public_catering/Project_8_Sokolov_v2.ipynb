### Шаг №1. Загрузите данные и подготовьте их к анализу

Загружаем библиотеки и данные

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns


%config InlineBackend.figure_format='retina'
%matplotlib inline
pd.options.display.max_columns = None
pd.options.display.max_rows = None
pd.options.display.max_colwidth = None
```

```python
try:
    rest = pd.read_csv(r'C:\Users\Иванов\rest_data.csv')    
except:
    rest = pd.read_csv('/datasets/rest_data.csv')
```

```python
display(rest.head())
rest.info()
print(f"Количество дубликатов в датафрейме: {rest.duplicated().sum()}")
```

Пропусков, дубликатов нет. Типы данных соответствуют.

```python
print(rest['object_name'].value_counts().head(20))
```

Переводим названия типов объектов в нижний регистр.

```python
rest['object_name'] = rest['object_name'].str.lower()
```

```python
print(rest['object_name'].value_counts())
```

Приводим названия заведений к одному типу.

```python
rest['object_name'] = rest['object_name'].replace('кафе «kfc»','kfc')
rest['object_name'] = rest['object_name'].replace('кафе «шоколадница»','шоколадница')
rest['object_name'] = rest['object_name'].replace('пиццерия «папа джонс»','папа джонс')
rest['object_name'] = rest['object_name'].replace('кафе «теремок»','теремок')
rest['object_name'] = rest['object_name'].replace('кофейня «шоколадница»','шоколадница')
rest['object_name'] = rest['object_name'].replace('ресторан «бургер кинг»','бургер кинг')
print(rest['object_name'].value_counts())
```

### Исследуйте соотношение видов объектов общественного питания по количеству. Постройте график.

```python
plt.figure(figsize=(15, 6))
plt.title("соотношение видов объектов общественного питания по количеству")
sns.histplot(y='object_type', data=rest)
plt.xlabel("Количество")
plt.ylabel("вид объекта")
plt.show()
```

Больше всего кафе. Количество кафетериев, закусочных и магазинов(отдел кулинарии) самое маленькое.

### Исследуйте соотношение сетевых и несетевых заведений по количеству. Постройте график.

```python
plt.figure(figsize=(12, 3))
plt.title("соотношение сетевых и несетевых заведений по количеству")
sns.histplot(y='chain', data=rest)
plt.xlabel("Количество")
plt.ylabel("Сетевое заведение")
plt.show()
```

Сетевых заведений меньше чем несетевых почти в 4 раза.

### Для какого вида объекта общественного питания характерно сетевое распространение?

```python
plt.figure(figsize=(15, 6))
plt.title("соотношение видов объектов общественного питания")
sns.histplot(y='object_type', data=rest, hue='chain')
plt.xlabel("Количество")
plt.ylabel("вид объекта")
plt.show()
```

Сетевое распространение характерно для предприятий быстрого обслуживания. А вот сетевых столовых и буфетов практически нет.

### Что характерно для сетевых заведений: много заведений с небольшим числом посадочных мест в каждом или мало заведений с большим количеством посадочных мест?


```python
chain_yes = rest.query('chain == "да"')
plt.figure(figsize=(15, 8))
plt.title("количество посадочных мест объектов общественного питания")
sns.scatterplot(x='number', y='object_type', data=rest, hue='chain')
plt.ylabel("вид объекта")
plt.xlabel("Количество мест")
plt.ylabel("вид объекта")
plt.show()
```

```python
rest_chain_yes = chain_yes.pivot_table(index=['object_name'], values='number', aggfunc=['count', 'mean']).reset_index()
rest_chain_yes.columns = ['object_name', 'count', 'number_mean']
rest_chain_yes = rest_chain_yes[rest_chain_yes['count'] > 1]
display(rest_chain_yes.head())
rest_chain_yes.describe()
```

Большим числом посадочных мест будем считать более 87. Небольшим - менее 20.

Большим количеством заведений в сети - более 7. Небольшим - 3 и менее.

```python
rest_chain_yes_1 = rest_chain_yes.query('number_mean>87 and count<=3')
rest_chain_yes_2 = rest_chain_yes.query('number_mean<20 and count>7')
rest_chain_yes_3 = rest_chain_yes.query('number_mean<20 and count<=3')
rest_chain_yes_4 = rest_chain_yes.query('number_mean>87 and count>7')
print(f"Количество заведений с большим числом посадочных мест и небольшим числом заведений в сети: {rest_chain_yes_1['number_mean'].count()}")
print(f"Количество заведений с небольшим числом посадочных мест и большим числом заведений в сети: {rest_chain_yes_2['number_mean'].count()}")
print(f"Количество заведений с небольшим числом посадочных мест и небольшим числом заведений в сети: {rest_chain_yes_3['number_mean'].count()}")
print(f"Количество заведений с большим числом посадочных мест и большим числом заведений в сети: {rest_chain_yes_4['number_mean'].count()}")
```

Для сетевых заведений характерно небольшое число заведений в сети с большим числом посадочных мест.

### Для каждого вида объекта общественного питания опишите среднее количество посадочных мест. Какой вид предоставляет в среднем самое большое количество посадочных мест? Постройте графики.


```python
plt.figure(figsize=(15, 6))
plt.title("среднее количество посадочных мест по видам объекта")
sns.barplot(x='number', y='object_type', data=rest)
plt.xlabel("Среднее подичество посадочных мест")
plt.ylabel("вид объекта")
plt.show()
```

В среднем самое большое количество посадочных мест в столовых и  ресторанах. Самое маленькое в магазинах, закусочных и кафетериях.

### Выделите в отдельный столбец информацию об улице из столбца address .


```python
rest['street'] = rest['address'].apply(lambda x: x.split(', ')[1])
```

```python
rest=rest[rest['street'].str.find('город') == -1]
rest=rest[rest['street'].str.find('поселение') == -1]
rest=rest[rest['street'].str.find('дом') == -1]
display(rest.head())
```

### Постройте график топ-10 улиц по количеству объектов общественного питания. Воспользуйтесь внешней информацией и ответьте на вопрос — в каких районах Москвы находятся эти улицы?


```python
rest_count = rest.pivot_table(index=['street'], values='object_type', aggfunc=['count']).reset_index()
rest_count.columns = ['street', 'count']
rest_top = rest_count.sort_values(by='count', ascending = False).head(10)

display(rest_top)
```

```python
plt.figure(figsize=(12, 3))
plt.title("топ-10 улиц по количеству объектов общественного питания")
sns.barplot(x='count', y='street', data=rest_top)
plt.xlabel("Количество")
plt.ylabel("Улица")
plt.show()
```

Лидирует проспект Мира.

```python
from io import BytesIO
import requests

spreadsheet_id = '1PIC4Z3cLfndG0urrEo1miqOXvXO5CrFL8aop5M5oahI'
file_name = 'https://docs.google.com/spreadsheets/d/{}/export?format=csv'.format(spreadsheet_id)
r = requests.get(file_name)
area = pd.read_csv(BytesIO(r.content))    
area.columns = ['street', 'areaid', 'okrug', 'area']
display(area.head())
```

```python
rest_top_pivot = rest_top.merge(area, how='left',  on='street', suffixes=(False, False))
```
